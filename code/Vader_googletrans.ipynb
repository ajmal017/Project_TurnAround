#### 원활한 번역을 위해 엑셀을 txt로 변환한뒤 번역을 해봄.

```python
from googletrans import Translator
translator = Translator()
text = ('안녕하세욤')
re = translator.translate(text, dest='en', src='auto')
print(re.text)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./Data_0.csv')
df = df['Title'].dropna()
```

```python
a = []
for i in df:
    a.append(i)
```

```python
with open('a.txt', 'w') as f:
    for line in a:
        f.write(line+'\n')
```

```python
with open('a2.txt', 'w') as f:
    f.writelines(a)
```

```python
a = 0
b = 100
c = 0
while c < len(df) :
    if b <= len(df):
        for i in range(a*100 , b):
            print(i)
        a += 1
        b += 100
        print(a, b)
    else:
        for i in range(a*100, len(df)):
            print(i)
        c = len(df)
```

```python
import time

a = 0
b = 100
c = 0

sentences = []

while c < len(df) :
    time.sleep(5)
    if b <= len(df):
        for i in range(a*100 , b):
            time.sleep(0.1)
            re = translator.translate(df[i], dest='en', src='auto')
            sen = re.text
            sentences.append(sen)
            print(re.text)
        a += 1
        b += 100
        print(a, b)
    else:
        for i in range(a*100, len(df)):
            time.sleep(0.1)
            re = translator.translate(df[i], dest='en', src='auto')
            sen = re.text
            sentences.append(sen)
            print(re.text)
        c = len(df)
```
