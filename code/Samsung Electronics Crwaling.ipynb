```python
import requests
from bs4 import BeautifulSoup
```

```python
request_headers = { 
'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
Safari/537.36'), } 

url = "https://finance.naver.com/item/board.nhn?code=005930&page=1"
html = requests.get(url,headers=request_headers ).text
soup = BeautifulSoup(html, 'html.parser')
a = soup.find_all('td',{'class':'title'})
```

```python
def crwaling(numpage):
    b = []
    for i in range(numpage+1):
        url = "https://finance.naver.com/item/board.nhn?code=005930&page=%s"%i
        requests.get(url,headers=request_headers ).text
        soup = BeautifulSoup(html, 'html.parser')
        a = soup.find_all('td',{'class':'title'})
        for i in a:
            b.append(i.find('a')['title'])
    return(b)
```

```python
samsung = crwaling(100)
```

```python
samsung
```

```python
f=open('samsung.txt','w')
f.write('\n'.join(samsung))
f.close()
```
