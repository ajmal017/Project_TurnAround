```python
# 먼저 pip install sentiment_lexicon , pip install vaderSentiment

from sentiment_lexicon import Lexicon
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(sentence):
    ss = sid.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(ss)) )
    
    
sentiment_analyzer_scores("bitcoin is gonna be raised because US economy is being well")
```

```python
from googletrans import Translator
translator = Translator()

def tr_en(sentence) :
    tr_results = translator.translate(sentence)
    return tr_results.text

def tr_jp(sentence) :
    tr_results = translator.translate(sentence, src='ko', dest='ja')
    return tr_results.text

def tr_jpen(sentence) :
    tr_results = translator.translate(sentence, src='ja', dest='en')
    return tr_results.text
```

```python
for i in data.head(100):
    print(sentiment_analyzer_scores(tr_jpen(tr_jp(i)))) # 번역 진행후 Score를 봐봄. 
```

```python
for i in data.head(1):
    ss = sid.polarity_scores(tr_jpen(tr_jp(i)))
    print(tr_jpen(tr_jp(i)))
    print(ss)
    print('pos :',ss['pos'])
    print('neg :',ss['neg'])
    print('neu :',ss['neu'])
    print('compound :',ss['compound'])
```

```python
import time
start = time.time()  # 시작 시간 저장

pos = []
neg = []
neu = []
compound = []
translation = []

for i in data:
    ss = sid.polarity_scores(tr_jpen(tr_jp(i)))
    translation.append(tr_jpen(tr_jp(i)))
    pos.append(ss['pos'])
    neg.append(ss['neg'])
    neu.append(ss['neu'])
    compound.append(ss['compound'])

df['translation'] = translation
df['pos'] = pos
df['neg'] = neg
df['neu'] = neu
df['compound'] = compound

print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
# 650개 401초 걸림. 6분~7분 코랩에서 돌릴것.
```

```python
df = df.sort_values(by=['compound'], ascending=False) # 오름차순으로 정렬
df.to_excel('./vader_test.xlsx')
```
