```python
# lg 화학 네이버 종목토론실 게시글 크롤링하기 

def crawling(numpage):
    res=[]
    for i in range(1,numpage+1):
        url = 'https://finance.naver.com/item/board.nhn?code=122870&page=%s'%i  # LG화학 
        request_headers = { 
    'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
    AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
    Safari/537.36'), } 
        source=requests.get(url,headers=request_headers)
        soup=BeautifulSoup(source.text,'html.parser')
        lists = soup.findAll('div', {'class' : 'section inner_sub'}) 
       
        for tmp in lists:
            name = []
            title_list = list(tmp.findAll('td', {'class': "title" }))
            
            for i in title_list:
                lst=i.find('a')['title']
                lst=lst.split('\n')
                name.extend(lst)
        res.append(name)
        
    return(res)

lg_chy=crawling(100)

with open('yg_네이버_종목토론실.txt','w') as f:
    for i in lg_chy:
        for j in i:
            f.write(j+'\n')
```
