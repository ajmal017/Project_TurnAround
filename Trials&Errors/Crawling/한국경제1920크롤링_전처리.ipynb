```
from konlpy.tag import Mecab

tokenizer = Mecab()

list_morph=[]
list_pos=[]
list_nouns=[]

df_morphs=pd.DataFrame()
df_nouns=pd.DataFrame()
df_pos=pd.DataFrame()

for i in range(len(clean_df)):
  list1 = clean_df.iloc[i,:].values.tolist()
  row1=[X_train.iloc[i,0]]   # 날짜(morph)
  row2=[X_train.iloc[i,0]]   # 날짜(nouns)
  row3=[X_train.iloc[i,0]]   # 날짜(pos)
  for text in list1:
    morphs_text = tokenizer.morphs(text)
    list_morph.append(morphs_text)
    
    nouns_text = tokenizer.nouns(text)
    list_nouns.append(nouns_text)
    
    pos_text = tokenizer.pos(text)
    list_pos.append(pos_text)
    
    row1.append(morphs_text)
    row2.append(nouns_text)
    row3.append(pos_text)

  row1=pd.Series(row1)
  df_morphs=pd.concat([df_morphs,row1],axis=1)

  row2=pd.Series(row2)
  df_nouns=pd.concat([df_nouns,row2],axis=1)

  row3=pd.Series(row3)
  df_pos=pd.concat([df_pos,row3],axis=1)


 
```

```
#df=df.to_numpy()
df_morphs=np.transpose(df_morphs)
df_morphs=pd.DataFrame(df_morphs)

df_nouns=np.transpose(df_nouns)
df_nouns=pd.DataFrame(df_nouns)

df_pos=np.transpose(df_pos)
df_pos=pd.DataFrame(df_pos)
```

```
# 19,20년의 휴일에 추가로 대체공휴일, 선거일, 근로자의날, 연말 등 krx거래소가 없는 휴일 데이터를 추가
holidays=list(holiday)
holi=[]
for i in holidays:
  i=i.strftime("%Y-%m-%d")
  holi.append(i)

worker_19=datetime.datetime(2019,5,1).strftime("%Y-%m-%d")   # 휴일리스트에 근로자의 날 추가
worker_20=datetime.datetime(2020,5,1).strftime("%Y-%m-%d")   # 근로자의 날 
alter_19=datetime.datetime(2019,5,6).strftime("%Y-%m-%d")   # 대체공휴일
alter_20=datetime.datetime(2020,1,27).strftime("%Y-%m-%d")  # 대체공휴일
elec_20=datetime.datetime(2020,4,15).strftime("%Y-%m-%d")  # 선거

offi_19=datetime.datetime(2019,12,31).strftime("%Y-%m-%d")
alter_20_2=datetime.datetime(2020,8,17).strftime("%Y-%m-%d")

holi.append(worker_19)
holi.append(worker_20)   # 2019년 5월 6일, 2020-01-27, 2020-04-15
holi.append(alter_19)
holi.append(alter_20)
holi.append(elec_20)
holi.append(offi_19)
holi.append(alter_20_2)

holi.sort()
```

```
# 라벨링
import copy

label_0=[]
label_025=[]
label_05=[]
label_075=[]


day_delta = datetime.timedelta(days=1)

sub1=0
sub2=0
sub3=0
cnt_0=0
cnt_holi=0

val_lst=[0, 0.25, 0.5, 0.75]

for v in val_lst:
  label=[]
  
  for i in range(train_data.shape[0]):
    if (train_data.iloc[i,2]=='금') | (train_data.iloc[i,2]=='토'):
      label.append(2)
      sub1+=1
      continue
    j=pd.to_datetime(train_data.iloc[i,0])
    y=int(j.strftime('%Y'))
    m=int(j.strftime('%m'))
    d=int(j.strftime('%d'))
    d_sub=datetime.date(y,m,d)  # train_data에 있는 날짜
    ds=d_sub+day_delta          # d_sub의 다음날 날짜
    q=ds.strftime("%Y-%m-%d") 
    #print(q)

    if (q in holi):
      sub1+=1
      label.append(2) 
      continue

    else:
      sub3+=1
      t1=kospi_data['년/월/일']==q
      t=kospi_data[t1]
      #print(t)
      val=t.iloc[0,3]
      if val>v:                 # v = [0, 0.25, 0.5, 0.75]
        label.append(1)
      elif val<-v:
        label.append(0)
      else:
        label.append(2)
  if v==0:
    label_0=label
  elif v==0.25:
    label_025=label
  elif v==0.5:
    label_05=label
  elif v==0.75:
    label_075=label
      
```
