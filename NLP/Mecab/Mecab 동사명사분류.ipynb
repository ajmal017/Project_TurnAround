```
from konlpy.tag import Mecab
```

```
tokenizer = Mecab()
```

```
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/문형님 파일/minus.csv')

list_none = []
list_morphs = []
list1 = df.loc[0][2:].dropna().tolist()
for i in list1:
  text = i
  nouns = tokenizer.nouns(text)
  morphs = tokenizer.pos(text)
  list_none.append(nouns)
  list_morphs.append(morphs)
```
