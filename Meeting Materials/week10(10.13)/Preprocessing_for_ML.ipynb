```python
import pandas as pd
from tqdm import tqdm
import re
from konlpy.tag import Okt
```

```python
data = pd.read_excel('./news.xlsx')
list3 = [] 
for i in range(len(data)) :
    list1 = data.loc[i].values.tolist()
    list2 = []
    for text in list1:
        temp = re.sub('[-=+,#/\?:^$.@*\"※~>`\'…》]', '', str(text))
        list2.append(temp)
    list3.append(list2)
clean_df = pd.DataFrame(list3)
clean_df

okt = Okt()
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']

# columns를 만든다. 
data2 = pd.read_excel("./train.xlsx")
list1 = list(data2['title'])
list2 = []

for text in list1:
  temp = re.sub('[-=+,#/\?:^$.@*\"※~>`\'…》]', '', text)
  list2.append(temp)
#list2

col_train = []

for i in tqdm(range(len(list2))):
  temp_X = []
  temp_X = okt.morphs(list2[i], stem=True) # 토큰화
  temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
  temp_X = [word for word in temp_X if len(word) > 1]
  col_train.append(temp_X)

#print(len(col_train))
col_train2 = list(set([y for x in col_train for y in x]))
#print(len(col_train2))

df = pd.DataFrame(columns = col_train2)
df['date'] = data['date']
df['labeling'] = data['labeling']

clean_df_columns = [columns for columns in clean_df.columns]
clean_df_columns.pop(0)
clean_df_columns.pop(0)
clean_df_columns.pop(0)
clean_df_columns.pop(0)
#clean_df_columns

df_columns = [df_column for df_column in df.columns]
df_columns.pop()
df_columns.pop()
#df_columns
    

for i in range(len(clean_df)):
    test = clean_df.loc[i].dropna().values.tolist()
    test.pop(0)
    test.pop(0)
    test.pop(0)
    test.pop(0)
    X_train = []
    for j in tqdm(range(len(test))) : # len test 기사수 
        temp_X = []
        temp_X = okt.morphs(test[j], stem=True) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        temp_X = [word for word in temp_X if len(word) > 1]
        X_train.append(temp_X)
    X_train = [y for x in X_train for y in x]  # 단어 1차원으로 피고 
    
    for df_column in tqdm(df_columns):
        df.loc[[i],[df_column]] = X_train.count(df_column)
     
```
